```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file
df = pd.read_csv('Companies.csv')

# Clean 'Employees' column by removing non-numeric characters
df['Employees'] = df['Employees'].str.replace('[^\d]', '', regex=True)

# Convert 'Employees' to int
df['Employees'] = df['Employees'].astype(int)

# Filter the DataFrame to include only companies with around 50,000 employees and located in California
filtered_df = df[(df['Employees'] >= 50000) & (df['Headquarters'].str.contains('California', case=False))]

# Plotting revenue for companies with around 50,000 employees in California
plt.figure(figsize=(10, 6))
sns.barplot(x='Revenue (USD millions)', y='Name', data=filtered_df, palette='Set2')
plt.title('Revenue of Companies with Around 50,000 Employees in California')
plt.xlabel('Revenue (USD millions)')
plt.ylabel('Company Name')
plt.show()
```

#LXML

```python
import requests
from lxml import etree
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
response = requests.get(url)
tree = etree.HTML(response.content)
```

```python
tree
```

```python
# Find the table
table = tree.xpath('//table[@class="wikitable sortable"]')[0]
table
```

```python

# Extract table headers
table_headers = [header.text.strip() for header in table.xpath('.//th')]
```

```python
table_headers
```

```python
df = pd.DataFrame(columns=table_headers)
```

```python
df
```

```python
# Extract table data
table_rows = table.xpath('.//tr')
for row in table_rows[1:]:
    row_data = []
    for data in row.xpath('.//td'):
        # Check if the data exists and extract the text content
        cell_content = data.xpath('string()').strip() if data.xpath('string()') else ''
        row_data.append(cell_content)
    df.loc[len(df)] = row_data
```

```python
df
```

```python
# Convert 'Revenue (USD millions)' column to numeric (float)
df['Revenue'] = df['Revenue'].str.replace('[^\d]', '', regex=True).astype(int)

# Convert 'Revenue growth' to float
df['Revenue growth'] = df['Revenue growth'].str.rstrip('%').astype(float)

# Clean 'Employees' column by removing non-numeric characters
df['Employees'] = df['Employees'].str.replace('[^\d]', '', regex=True).astype(int)

# Confirm the data types after conversion
print(df.dtypes)






# Export to CSV
df.to_csv('Companies_lxml_fixed.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv('Companies_lxml_fixed.csv')

# Convert 'Revenue (USD millions)' column to numeric (float)
df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')

# Drop rows with missing revenue values
df.dropna(subset=['Revenue'], inplace=True)

# Sort the DataFrame by revenue in descending order
df.sort_values(by='Revenue', ascending=False, inplace=True)

# Plotting line spacing of Revenue for top 10 companies
plt.figure(figsize=(10, 6))
plt.plot(df['Name'].head(10), df['Revenue'].head(10), marker='o', linestyle='--')
plt.title('Line Spacing of Revenue for Top 10 Companies')
plt.xlabel('Company Name')
plt.ylabel('Revenue (USD millions)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
# Plotting curves of Revenue Growth for top 10 companies
plt.figure(figsize=(10, 6))
plt.plot(df['Name'].head(10), df['Revenue growth'].head(10), marker='x', linestyle='-', color='green')
plt.title('Curves of Revenue Growth for Top 10 Companies')
plt.xlabel('Company Name')
plt.ylabel('Revenue Growth')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
